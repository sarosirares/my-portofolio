```python
weather = pd.read_csv("C:\\Facultate\\Facultate\\Anul 3\\Semestrul 1\\KBS\\Lab\\Lab 3\\Summary of Weather.csv")
print(weather.head())
weather.head(2)
```

```python
# Cleaning the Aerial Bombing Data

operations.dropna(subset=['Country', 'Target Longitude', 'Takeoff Longitude', 
                          'Takeoff Latitude', 'Target Latitude'], inplace=True)

# Convert Mission Date to datetime for time series analysis
operations['Mission Date'] = pd.to_datetime(operations['Mission Date'])

# Cleaning the Weather Condition Data (Column Selection)

columns_loc = ["WBAN", "NAME", "STATE/COUNTRY ID", "Latitude", "Longitude"]
wstations = wstations.loc[:, columns_loc].rename(columns={'STATE/COUNTRY ID': 'Country_ID'})

columns_weather = ["STA", "Date", "MeanTemp"]
weather = weather.loc[:, columns_weather].rename(columns={'STA': 'WBAN'})

# Convert Date to datetime for time series analysis
weather['Date'] = pd.to_datetime(weather['Date'])

print("weather after cleaning:")
print(weather.info(verbose=False, show_counts=True))
```

```python
# f. Time Series Analysis: MeanTemp for a single station
station_id = weather['WBAN'].iloc[0] # Select the first station for analysis (WBAN 10001)
single_station = weather[weather['WBAN'] == station_id].set_index('Date').sort_index()

plt.figure(figsize=(15, 6))
plt.plot(single_station.index, single_station['MeanTemp'], color='tab:orange', linewidth=1)
plt.title(f'Mean Temperature Over Time for Weather Station {station_id}')
plt.xlabel('Date')
plt.ylabel('Mean Temperature ($^\circ$C)')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# --- Data Preparation (Assuming df_weather is loaded and cleaned) ---

station_id = weather['WBAN'].iloc[0] 
df_single_station = weather[weather['WBAN'] == station_id].set_index('Date').sort_index()
ts = df_single_station['MeanTemp'].dropna() # Time Series object

print(f"--- Time Series Analysis for Station {station_id} (MeanTemp) ---")

period_val = 30 

multiplicative_decomposition = seasonal_decompose(ts, model='multiplicative', period=period_val)
plt.rcParams.update({'figure.figsize': (10, 8)})
multiplicative_decomposition.plot().suptitle('Multiplicative Decomposition', fontsize=16)

# 2. Testing for Stationarity (ADF Test)

adf_result = adfuller(ts)
print("\nAugmented Dickey-Fuller (ADF) Test Results:")
print(f"Test Statistic: {adf_result[0]:.4f}")
print(f"P-value: {adf_result[1]:.4f}")
print(f"Critical Values (for rejection): {adf_result[4]}")
print("\nConclusion: If P-value > 0.05, the series is likely non-stationary.")

# 3. Autocorrelation (ACF) and Partial Autocorrelation (PACF) Plots

fig, axes = plt.subplots(1, 2, figsize=(12, 4), dpi=100)
plot_acf(ts, lags=60, ax=axes[0], title='Autocorrelation Function (ACF)')
plot_pacf(ts, lags=60, ax=axes[1], title='Partial Autocorrelation Function (PACF)')
```
